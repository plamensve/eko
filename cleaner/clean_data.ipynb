```python
import pandas as pd
```

```python
data = pd.read_excel('dd.xlsx')
```

```python
def load_cards(dataset):

    df = dataset.copy()

    df = df.rename(columns={
        "пуснати карти за издаване списък": "Company",
        "РЕГ. НОМЕР МПС / Vehicle registration number": "Name",
        "Номер карта": "Number"
    })

    df = df[["Company", "Name", "Number"]]

    df = df[df["Name"].notna()]

    # премахва празните места в регистрационния номер
    df["Name"] = df["Name"].astype(str).str.replace(" ", "")

    df["Number"] = (
        df["Number"]
        .astype(str)
        .str.replace(r"\D", "", regex=True)
    )

    df = df[df["Number"].str.len() > 0]

    return df
```

```python
load_cards(data)
```

```python
load_cards(data).to_excel("cards.xlsx", index=False)
```
